# Power Spectral Densities

Are very important function in seismology! Below is a simple example how ObPsy allows calculating PPSDs for a seismic trace. PPSDs are also used to characterize the noise floor of ground motion instruments (e.g., seismometers, accelerometers, rotation sensors). 

The goal here is to understand the concept of Low and High Noise models for our Earth, and the meaning of the values and axes in the graph!

**Exercises:**

- Check the ObsPY documentation for parameters of the ppsd and modify the plot

```python
from obspy import read
from obspy.signal import PPSD
from obspy.io.xseed import Parser


st = read("https://examples.obspy.org/BW.KW1..EHZ.D.2011.037")
parser = Parser("https://examples.obspy.org/dataless.seed.BW_KW1")
ppsd = PPSD(st[0].stats, metadata=parser)
ppsd.add(st)

st = read("https://examples.obspy.org/BW.KW1..EHZ.D.2011.038")
ppsd.add(st)

ppsd.plot(cumulative=True)

```
