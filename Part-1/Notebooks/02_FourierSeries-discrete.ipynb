## Fourier Series (discrete)

#### Example

In the following example the function 

$f(x) = x^2 $

is approximated by a discrete Fourier series in the interval [0,  2 $pi$].

As a reference, the **continuous** transform is given by

$$
x^2 \sim \frac{4\pi^2}{3} + \sum_{n=1}^\infty \frac{4}{n^2} \cos(nx)
$$

But this is not used here. In this example we generate the discrete Fourier series using the functional values at the regular spaced grid points. 

First, explore the code and relate the code lines with the original equations for the discrete Fourier transform! Try to understand the relevance of the approximation order N and the exact interpolation property at the grid points (plotted with markers "+"). 

Programming exercise at the bottom. 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Program to illustrate discrete Fourier series
# H. Igel, 7/23

fs=11

# Functions will be considered in the interval [0 2pi]

# Global parameters
N = 64     # number of intervals and number of base functions (even)
n=int(N/2)
nx = 1000  # grid points for approximate function


# Initialize regular space grid
xi = np.linspace(0,2*np.pi,N+1)

# Initialize function to approximate
fi = xi*xi    # f(x) = x^2

# Calculate Fourier coefficients

# a_k (cosines)
a=np.zeros(N)
a0=2./N*np.sum(fi)
for k in range(1,n+1):
    for i in range(1,N+1):
        a[k-1]=a[k-1]+2/N*fi[i]*np.cos(k*xi[i])
        
        
# b_k (sines)
b=np.zeros(N);
for k in range(1,n+1):
    for i in range(1,N+1):
        b[k-1]=b[k-1]+2/N*fi[i]*np.sin(k*xi[i])

        
# Assemble Fourier series

# Initialize space (fine meshing)
#x=0:.001:2*pi;
#x  = np.linspace(0,2*np.pi,6284)
x = np.arange(0,2*np.pi,0.001)

# Initialize approximate function
g=x * 0

for k in range(0,n-1):
    g = g + a[k]*np.cos((k+1)*x) + b[k]*np.sin((k+1)*x)

g = g + a0/2 + a[n-1]/2*np.cos((n)*x);

plt.plot(x,g,linestyle='solid')
plt.plot(xi,fi,marker='+')
```
